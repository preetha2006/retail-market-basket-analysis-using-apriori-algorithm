```python
transactions = data.groupby('Member_number')['itemDescription'].apply(list)
transactions.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df_encoded = pd.DataFrame(te_array, columns=te.columns_)
df_encoded.head()
```

```python
frequent_itemsets = apriori(df_encoded,min_support=0.01,use_colnames=True)
frequent_itemsets.sort_values(by="support", ascending=False).head()
```

```python
frequent_itemsets['itemset_length'] = frequent_itemsets['itemsets'].apply(len)

plt.figure()
sns.countplot(x='itemset_length', data=frequent_itemsets)
plt.title("Distribution of Frequent Itemset Sizes")
plt.xlabel("Itemset Size")
plt.ylabel("Count")
plt.show()
```

```python
rules = association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3)
rules.head()
```

```python
rules.sort_values(by="lift", ascending=False).head(10)
```

```python
plt.figure()
plt.scatter(rules['support'], rules['confidence'])
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title("Support vs Confidence")
plt.show()
```

```python
plt.figure()
plt.hist(rules['lift'], bins=30)
plt.xlabel("Lift")
plt.ylabel("Frequency")
plt.title("Lift Distribution")
plt.show()
```

```python
strong_rules = rules[(rules['confidence'] >= 0.6) &(rules['lift'] >= 1.2)]
strong_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head()
```
